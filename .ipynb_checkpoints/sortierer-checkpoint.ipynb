```python
path = 'C:/Users/User/Desktop/Muell_and_clean'
path_cyg = '/cygdrive/c/Users/User/Desktop/Muell_and_clean'
files_all_original_size = !ls {path_cyg}/all_original_size
anzahl_original_size = len(files_all_original_size)
w_size = 500 #window size

while 1:
    method = input('resize or cut?')
    if method == 'resize':
        break
    elif method == 'cut':
        break
    else:
        continue
        
if anzahl_original_size > 0:
    for i in range(0, anzahl_original_size):   
        with Image.open(path + '/all_original_size/' + files_all_original_size[i]) as file:
            
            if method == 'cut':
                img = np.array(file) 
                x,y,z = np.shape(img)
                img_size = img[int((x-w_size)/2):int((x+w_size)/2), int((y-w_size)/2):int((y+w_size)/2)]
                img_size_uint8 = Image.fromarray(np.uint8(img_size))
                img_size_uint8.save(path + '/all/' + files_all_original_size[i])
                
            elif method == 'resize':
                img = file
                if img.mode != "RGB":
                    img = img.convert("RGB")
                img_size = img.resize((w_size, w_size)) 
                img_size.save(path + '/all/' + files_all_original_size[i])
            
```

```python
def zf(klasse, i):
    shutil.copy2(path + '/all/' + files_all[i], path + klasse)
    shutil.move(path + '/all/' + files_all[i], path + '/zugeschnitten_und_sortiert')
    if method == 'resize':
        os.rename(path + klasse + '/' + files_all[i], path + klasse + klasse + '_' + time.strftime('%Y%m%d') + '_' + f'{i:06}' + '.jpg')
    elif method == 'cut':
        os.rename(path + klasse + '/' + files_all[i], path + klasse + klasse + '_' + files_all[i])
    plt.close('all')
    cv2.destroyAllWindows()
    return i+1
```

```python
clean = 'a'
muell = 'd'
scheisse = 't'
ende = 'q'
files_all = !ls {path_cyg}/all
anzahl_images = len(files_all)

dummy_cat = cv2.imread('dummy_cat.jpg') # load a dummy image
i = 0
plt.figure('clean or muell')

while 1:
    if i < anzahl_images:  
        plt.imshow(Image.open(path + '/all/' + files_all[i]))
        cv2.imshow('dummy cat', dummy_cat)
        
    else:
        plt.close('all')
        cv2.destroyAllWindows()
        print(i)
        break
        
    k = cv2.waitKey(33)
    
    if k == 27: # Esc key to stop
        plt.close('all')
        cv2.destroyAllWindows()
        break
        
    elif k == -1:  # normally -1 returned,so don't print it
        continue
        
    elif k == ord(muell):
        i = zf('/muell', i)
        
    elif k == ord(clean):
        i = zf('/clean', i)
        
    elif k == ord(scheisse):
        i = zf('/scheisse', i)
        
    elif k == ord(ende):
        plt.close('all')
        cv2.destroyAllWindows()
        print(i)
        break
```
